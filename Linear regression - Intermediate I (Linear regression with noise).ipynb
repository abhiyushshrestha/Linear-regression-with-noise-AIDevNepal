```python
import numpy as np #importing numpy with alias 'np' for using numpy based array which are optimized for calculations
import matplotlib.pyplot as plt #importing matplotlib library with alias 'plt' which is used for plotting the graph
from sklearn.linear_model import LinearRegression #importing sklearn library that is used for linear regression

```

### Creating a training data sets : Input data
#### x as features or independent variable
#### y as label or dependent variable

```python
x = np.linspace(-1,1,100) #Creating equal space 101 points
```

```python
print(x)
```

```python
plt.plot(x)
```

```python
noise = np.random.randn(x.shape[0])*0.33 #Creating noise to randomize the data
```

```python
print(noise)
```

```python
plt.plot(noise)
```

```python
y = 2 * x + noise #y = mx + c
```

```python
print(y)
```

```python
plt.plot(y)
```

### Converting rank 1 array to 1d column vector


```python
x_col_1d = x.reshape(-1,1)
y_col_1d = y.reshape(-1,1)
noise_col_1d = noise.reshape(-1,1)
```

```python
print("Shape of x: ",x_col_1d.shape)
print("Shape of y: ",y_col_1d.shape)
print("Shape of noise: ", noise_col_1d.shape)
```

### Simplifying the variable names


```python
X = x_col_1d
Y = y_col_1d
Noise = noise_col_1d
```

```python
plt.plot(Y)
plt.show()
```

### Creating Linear regression object


```python
model = LinearRegression()
```

### Training a model with datasets X and Y

```python
model.fit(X,Y)
```

### Predicting a value


```python
model.predict(10)
```

### Calculated value of slope and intercept

```python
m = model.coef_
c = model.intercept_
print("Slope(m): ", m)
print("Intercept(c): ", c)
```

```python
plt.plot(X,Y)
plt.plot(X, m * X + c, 'r')
```

```python
model.score(X,Y)
```
